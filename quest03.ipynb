```python
import pandas as pd
import numpy as np
import json
```

```python
#importando o arquivo em JSON

with open('dados.json') as json_file:
    datos = json.load(json_file)
```

```python
#Carregando o dataframe
dados = pd.DataFrame(datos)
dados.head()
```

```python
#Tratamento dos dados

print(dados.isna().sum())

#sem valores nulos
```

```python
#Procurando faturamento zero
fatu_zero = dados[dados['valor'] == 0]
```

```python
#Dias com faturamento zero
print('Dias de faturamento zero: ')
print(fatu_zero)
```

```python
#Dias com faturamento maior que 0
fatu_maior = dados[dados['valor'] > 0]
```

```python
#Dias com fatu maior que 0
print('Dias com faturamento maior que zero: ')
print(fatu_maior)
```

```python
#Encontrando o valor mínimo(Excluindo o zero)
valor_minimo = fatu_maior['valor'].min()
print('-*'*20)
print(f'Valor mínimo de faturamento: {valor_minimo}')
print('-*'*20)
```

```python
#Entrando o valor maximo (Excluindo o zero)
valor_maximo = fatu_maior['valor'].max()
print('-*'*20)
print(f'Valor máximo de faturamento: {valor_maximo}')
print('-*'*20)
```

```python
#Calcular a média de faturamento diario(Excluíndo os valores zero)
media_fatu = fatu_maior['valor'].mean()
print('-+'*20)
print(f'Média de faturamento: {media_fatu}')
print('-+'*20)
```

```python
#Indentificar os dia com faturamento acima da média

dias_acima_media = fatu_maior[fatu_maior['valor'] > media_fatu]
print('-*'*20)
print('Dias com faturamento acima da média: ')
print(dias_acima_media)
print('-*'*20)
```

Criar um DataFrame com os resultados


```python
resultados = pd.DataFrame({
    "Descrição": ["Valor mínimo", "Valor máximo", "Média de faturamento",],
    "Valor": [valor_minimo, valor_maximo, media_fatu]
})

print(resultados)
```
